```python
# Part 1
import matplotlib.pyplot as plt
```

```python
with open('1200wsj.txt') as file:
    lines = file.read().strip()
```

```python
chars = {}
for char in lines:
    newChar = char.lower()
    if newChar in chars:
        chars[newChar] += 1
    else:
        chars[newChar] = 1
```

```python
sortedChars = sorted(chars.items(), key=lambda item: item[1], reverse=True)   

characters = []
counts = []

for charCount in sortedChars:
    characters.append(charCount[0])
    counts.append(charCount[1])
    
plt.bar(characters, counts)
```

```python
# Extra: ust to show the top 20. Besides the space, e is the most frequently used character
plt.bar(characters[0:20], counts[0:20])
```

```python
tokens = lines.split()
wordcounts = {}
for token in tokens:
    if token in wordcounts:
        wordcounts[token] += 1
    else:
        wordcounts[token] = 1
```

```python
# all converted to lowercase
wordcountsLower = {}
for word, count in wordcounts.items():
    newWord = word.lower()
    wordcountsLower[newWord] = count
wordcountsLower
```

```python
words = []
counts = []
for lowerWord, lowerCount in wordcountsLower.items():
    words.append(lowerWord)
    counts.append(lowerCount)
```

```python
suffixDict = {}

for word, count in wordcountsLower.items():
    if len(word) < 4:
        continue
    
    suffix2 = word[-2:]
    suffix2word = word[:-2]
    if suffix2word in words:
        if suffix2 in suffixDict:
            suffixDict[suffix2] += 1
        else:
            suffixDict[suffix2] = 1
            
    suffix3 = word[-3:]
    suffix3word = word[:-3]
    if suffix3word in words:
        if suffix3 in suffixDict:
            suffixDict[suffix3] += 1
        else:
            suffixDict[suffix3] = 1
```

```python
sortedSuffix = dict(sorted(suffixDict.items(), key=lambda item: item[1], reverse = True))
```

```python
sortedSuffix
```

```python
top20 = dict(list(sortedSuffix.items())[:20])
```
